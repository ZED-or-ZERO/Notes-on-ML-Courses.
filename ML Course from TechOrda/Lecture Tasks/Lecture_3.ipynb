```python
# Task 1
numbers = [] 
count = 0
with open('../data/numbers.txt', 'r') as f:
    for line in f:
        line = line.strip('\n')
        count += int(line) 
        numbers.append(int(line))

print(f"Count numbers: {count}")
print(numbers)
```

```python
# 2 
cube = [x**3 for x in numbers if x % 3 != 0]
cube
```

```python
# 5 
import json 

file_name = '../data/student_grades.json'

with open(file_name, 'w', encoding='utf-8') as f:
    json.dump(students, f)
```

```python
with open(file_name, 'r', encoding='utf-8') as f:
    loaded_data = json.load(f)

all_scores = list(loaded_data.values())
```

```python
total_sum = sum(all_scores)
count = len(all_scores)
average_score = total_sum / count
```

```python
print(f"Average score : {average_score:.2f}")
```
