### Кейс-тапсырма: "Студенттік Грант Калькуляторы"


Тапсырма 1: Деректерді  сақтау және деректер типтері (айнымалылар)


```python
# 1. Айнымалыларды анықтау және мән беру
student_name = "Айжан Қанатқызы"  # str типі
student_gpa = 3.85               # float типі
student_course = 2               # int типі
special_status = True            # bool типі (Осал санат)

# Барлық айнымалылардың типтерін экранға шығару
print("--- Тапсырма 1: Деректер Типтері ---")
print(f"Айнымалы {student_name=}: Типі - {type(student_name)}")
print(f"Айнымалы {student_gpa=}: Типі - {type(student_gpa)}")
print(f"Айнымалы {student_course=}: Типі - {type(student_course)}")
print(f"Айнымалы {special_status=}: Типі - {type(special_status)}")
print("-" * 40)

# Базалық грантты сақтау үшін айнымалыны бастапқы мәнмен анықтаймыз
base_grant = 0
```

Тапсырма 2: Шартты Логика (Гранттың Базалық Мөлшерін Анықтау)

```python
# 2. Шартты Логика (Базалық Грант)
print("--- Тапсырма 2: Базалық Грантты Анықтау ---")
if student_gpa >= 3.5:
    base_grant = 50000  # Жоғары грант
    print(f"GPA ({student_gpa}) 3.5-тен жоғары. Базалық грант: {base_grant} теңге (Жоғары).")
elif 3.0 <= student_gpa < 3.5:
    base_grant = 25000  # Орташа грант
    print(f"GPA ({student_gpa}) 3.0 пен 3.5 арасында. Базалық грант: {base_grant} теңге (Орташа).")
else:
    base_grant = 0      # Грант жоқ
    print(f"GPA ({student_gpa}) 3.0-ден төмен. Базалық грант: {base_grant} теңге (Грант жоқ).")
    
# Жалпы грант сомасын есептеу үшін айнымалыны базалық грантпен бастаймыз
total_grant = base_grant
print("-" * 40)
```

Тапсырма 3: Цикл. Олимпиадалар

```python
# 3. Циклдар және Бонус
print("--- Тапсырма 3: Циклдар және Олимпиада Бонусы ---")

# Олимпиадалар тізімі
olympiad_bonuses = ["Математика", "Физика", "Тарих", "Ағылшын тілі", "Информатика"]
olympiad_participation_count = 3 # Студент қатысқан олимпиадалар саны (шартты тексеру үшін)

# 3.1. Олимпиадаға қатысқандығы үшін бонус
if olympiad_participation_count >= 3:
    bonus_amount = 5000
    total_grant += bonus_amount
    print(f"Студент {olympiad_participation_count} олимпиадаға қатысты. Бонус: {bonus_amount} теңге қосылды.")
else:
    print("Олимпиадаға қатысу бонусын алу үшін кем дегенде 3 олимпиадаға қатысу қажет.")

# 3.2. for циклімен олимпиада тізімін шығару
print("\nСтудент қатысуы мүмкін олимпиадалар тізімі:")
for olympiad in olympiad_bonuses:
    print(f"- {olympiad}")

# 3.3. while циклімен ескерту беру
if base_grant == 0:
    print("\n! Грант Жоқ Ескертуі (While циклі):")
    warning_count = 0
    while warning_count < 3:
        print("Грант алу үшін GPA-ды көтеру керек!")
        warning_count += 1

print("-" * 40)
```

Тапсырма 4: Функцияны Қолдану. (Әлеуметтік Жеңілдік)

```python
# 4. Функцияны Қолдану (Әлеуметтік Жеңілдік)
print("--- Тапсырма 4: Әлеуметтік Жеңілдік Функциясы ---")

# Әлеуметтік жеңілдіктерді есептейтін функция
def social_bonus(special_status: bool) -> int:
    """
    Студенттің әлеуметтік мәртебесіне байланысты бонус мөлшерін есептейді.
    """
    if special_status:
        return 10000
    else:
        return 0

# Функцияны шақыру және нәтижесін жалпы грантқа қосу
social_grant_amount = social_bonus(special_status)
total_grant += social_grant_amount

print(f"Әлеуметтік мәртебеге байланысты бонус: {social_grant_amount} теңге.")
print("-" * 40)
```

Тапсырма 5: Жалпы Грантты Есептеу

```python
# 5. Жалпы Грантты Есептеу және Нәтиже
print("--- Тапсырма 5: Жалпы Гранттың Қорытындысы ---")

# Жалпы грантты шығару
print(f"Студент **{student_name}** үшін жалпы грант мөлшері: **{total_grant} теңге**.")
print("-" * 40)
```
