## 1-тапсырма: Токенизация (Tokenization)

**Теория:**
Токенизация  мәтінді модель түсінетін кішігірім бөліктерге (токендерге) бөлу процесі. Бұл NLP-дегі ең алғашқы қадам. Трансформер модельдері шикі мәтінді қабылдамайды, олар сандар тізбегін қабылдайды. `AutoTokenizer` класы мәтінді сандарға айналдыруға көмектеседі.

**Тапсырма:**
`bert-base-uncased` моделі үшін токенизаторды жүктеп, берілген сөйлемді токенизациялаңыз (яғни, input IDs және attention mask алыңыз).

```python
from transformers import AutoTokenizer

sentence = "Natural Language Processing is amazing and fun to learn."

# 1. 'bert-base-uncased' моделінің токенизаторын жүктеңіз
# 2. sentence айнымалысын токенизатор арқылы өңдеп, нәтижені 'inputs' айнымалысына сақтаңыз
# 3. Нәтижені экранға шығарыңыз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

inputs =  tokenizer(sentence, padding=True, truncation = True, return_tensors='pt')

print(inputs)
```

```python
print(inputs)
```


## 2-тапсырма: Мәтінді классификациялау (Text Classification)

**Теория:**
Мәтінді классификациялау — мәтінге белгілі бір санатты немесе белгіні тағайындау (мысалы, спам/спам емес, позитивті/негативті). Hugging Face-те `pipeline()` функциясы алдын ала үйретілген модельдерді оңай қолдануға мүмкіндік береді. "sentiment-analysis" (сезімді талдау) — классификацияның кең таралған түрі.

**Тапсырма:**
`pipeline` функциясын пайдаланып, "sentiment-analysis" тапсырмасын орындаңыз. Берілген мәтіндер тізімінің эмоционалдық бояуын (позитивті немесе негативті) анықтаңыз.

```python
from transformers import pipeline
```

```python
from transformers import pipeline

texts = [
    "I love using Python for data science!",
    "The weather is terrible today, I hate it."
]

# 1. 'sentiment-analysis' үшін pipeline құрыңыз
# 2. 'texts' тізімін pipeline-ға беріп, нәтижесін алыңыз
# 3. Әр мәтіннің label (белгісі) мен score (сенімділік деңгейін) экранға шығарыңыз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ

classifer = pipeline('sentiment-analysis')

results = classifer(texts)

print(f"\nText: {texts}")

for text, results in zip(texts, results):
    print(f"Label: {results['label']}, Score: {results['score']:.6f}")
```


## 3-тапсырма: Атаулы нысандарды шығару (NER)

**Теория:**
Named Entity Recognition (NER) мәтіннен адам аттарын (PER), ұйымдарды (ORG), локацияларды (LOC) және басқа да маңызды нысандарды тауып, оларды классификациялау. Бұл ақпаратты іздеу жүйелерінде жиі қолданылады.

**Тапсырма:**
Төмендегі мәтіннен атаулы нысандарды анықтау үшін `ner` пайплайнын қолданыңыз. Нәтижелерді топтастыру үшін `aggregation_strategy="simple"` параметрін қосуды ұмытпаңыз.

```python
ner_text = "Elon Musk is the CEO of SpaceX, located in California."

# 1. NER тапсырмасы үшін pipeline құрыңыз (aggregation_strategy="simple" қолданыңыз)
# 2. 'ner_text' айнымалысын модельге беріңіз
# 3. Табылған нысандарды (entity, score, word) экранға шығарыңыз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ

ner_pipeline = pipeline("ner", aggregation_strategy='simple')

entities = ner_pipeline(ner_text)

for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Score: {entity['score']}")
```



## 4-тапсырма: Мәтінді қысқарту (Summarization)

**Теория:**
Summarization үлкен мәтінді оның негізгі мағынасын жоғалтпай қысқаша нұсқаға айналдыру. Бұның екі түрі бар:

1. **Extractive:** Мәтіннен маңызды сөйлемдерді сол күйінде қиып алады.
2. **Abstractive:** Мәтіннің мазмұнын түсініп, жаңа сөздермен қысқаша баяндайды (адам сияқты). Трансформерлер (мысалы, BART, T5) көбінесе екінші әдісті қолданады.

**Тапсырма:**
Берілген ұзақ мәтінді `summarization` пайплайны көмегімен қысқартыңыз. `min_length` және `max_length` параметрлерін қолданыңыз.

```python
long_text = """
The Transformer model architecture, introduced in the paper "Attention Is All You Need",
has revolutionized the field of Natural Language Processing. Unlike previous Recurrent Neural Networks (RNNs),
Transformers process the entire input sequence in parallel, allowing for much faster training and
better handling of long-range dependencies in text. This architecture is the foundation of
modern models like BERT, GPT, and T5.
"""

"""    
Transformers process the entire input sequence in parallel, allowing for much faster training and better handling 
of long-range dependencies in text . This architecture is the foundation of modern models like BERT, GPT, and T5 .
"""

# 1. 'summarization' үшін pipeline жүктеңіз
# 2. Мәтінді қысқартыңыз (max_length=50, min_length=20 деп орнатыңыз)
# 3. Қысқартылған мәтінді экранға шығарыңыз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ


from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
model = AutoModelForSeq2SeqLM.from_pretrained("sshleifer/distilbart-cnn-12-6")

inputs = tokenizer(long_text, return_tensors="pt", truncation=True)
summary_ids = model.generate(inputs["input_ids"], max_length=50, min_length=20)
print(tokenizer.decode(summary_ids[0], skip_special_tokens=True))
```

## 5-тапсырма: Машиналық аударма (Translation)

**Теория:**
Машиналық аударма (Machine Translation) мәтінді бір тілден екінші тілге автоматты түрде аудару. Hugging Face-те `Helsinki-NLP` немесе `T5` сияқты модельдер көптеген тіл жұптарын қолдайды.

**Тапсырма:**
Ағылшын тілінен француз тіліне (немесе басқа қалаған тіліңізге) аударатын пайплайн құрыңыз. `task` ретінде `"translation_en_to_fr"` қолдануға болады немесе арнайы модельді (мысалы, `Helsinki-NLP/opus-mt-en-fr`) көрсетуге болады.

```python
text_to_translate = "Hugging Face makes deep learning accessible to everyone."

# 1. Аударуға арналған pipeline құрыңыз (мысалы, "translation_en_to_fr")
# 2. Мәтінді аударыңыз
# 3. Аударма нәтижесін көрсетіңіз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ

translator = pipeline(model="Helsinki-NLP/opus-mt-en-fr")
result = translator(text_to_translate)
print(result)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-en-fr"

text_to_translate = "Hugging Face makes deep learning accessible to everyone."

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

inputs = tokenizer(text_to_translate, return_tensors="pt")

outputs = model.generate(**inputs)

translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Аударма нәтижесі:", translated_text)
```

## 6-тапсырма: Қазақ тіліндегі модельмен жұмыс (Masked Language Modeling)

**Теория:**
Қазақ тіліне арналып жасалған (pre-trained) модельдердің бірі — `Amirlik/kaz-roberta-bert-base`. Бұл модель RoBERTa архитектурасына негізделген және қазақ тіліндегі үлкен мәтіндік корпустарда (Wikipedia, жаңалықтар) оқытылған.
**Masked Language Modeling (MLM)** — бұл модельге сөйлемнің бір сөзін жасырып (`<mask>`), сол сөзді контекст арқылы табуды тапсыру. Бұл модельдің тілді қаншалықты терең түсінетінін тексеретін тапсырма.

**Тапсырма:**
`xlm-roberta-base` моделін және `fill-mask` пайплайнын қолданып, сөйлемдегі түсіп қалған сөзді табыңыз.

```python
from transformers import pipeline

# Біз қолданатын қазақстандық модель
model_name = "xlm-roberta-base"
sentence = "Қазақстанның астанасы <mask> қаласы"

# 1. 'fill-mask' тапсырмасы үшін pipeline құрыңыз және жоғарыдағы 'model_name' моделін көрсетіңіз
# 2. 'sentence' айнымалысын пайплайнға беріп, нәтижені алыңыз
# 3. Модель ұсынған ең ықтимал сөзді (token_str) және оның сәйкестік ұпайын (score) экранға шығарыңыз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ

fill_mask = pipeline('fill-mask', model=model_name)
result = fill_mask(sentence)
```

```python
for res in result:
    print(f"Word: {res['token_str']}, Score: {res['score']}")
```

## 7-тапсырма: Қазақ тіліндегі модельді Fine-tuning-ке дайындау

**Теория:**
Дайын модельді (мысалы, `xlm-roberta-base`) өз деректерімізге бейімдеу үшін оны жүктеп, оқыту параметрлерін баптау қажет. Бұл тапсырмада сіз модельді және токенизаторды жүктеп, `TrainingArguments` арқылы оқыту процесінің конфигурациясын жасайсыз.

**Тапсырма:**
`xlm-roberta-base` моделін `AutoModelForSequenceClassification` класы арқылы жүктеңіз (мысалы, 2 класс үшін). Содан кейін оқыту параметрлерін баптаңыз.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments

model_name = "xlm-roberta-base"

# 1. Токенизаторды 'model_name' атауы бойынша жүктеңіз (AutoTokenizer қолданыңыз)
# 2. Модельді классификация тапсырмасы үшін жүктеңіз (num_labels=2 деп көрсетіңіз)
# 3. TrainingArguments объектісін құрыңыз:
#    - output_dir параметрін "./kaz_bert_finetuned" деп қойыңыз
#    - learning_rate параметрін 2e-5 деп қойыңызto
#    - num_train_epochs параметрін 3 деп белгілеңіз
# 4. Модельдің архитектурасын немесе аргументтерді экранға шығарып тексеріңіз

# КОДТЫ ОСЫ ЖЕРГЕ ЖАЗЫҢЫЗ

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

training_args = TrainingArguments(
    output_dir="./kaz_bert_fine_tuned",
    learning_rate = 2e-5,
    num_train_epochs = 3, 
)

print("Configuration Complete!")
```
