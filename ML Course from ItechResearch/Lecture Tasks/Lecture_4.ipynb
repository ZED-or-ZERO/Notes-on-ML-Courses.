## # 1 Create Class

```python
class Rectangle:
    def __init__(self, x: int, y: int):
        self.x = x
        self.y = y 

    def square(self):
        return self.x * self.y 
    
    def perimeter(self):
        return (self.x + self.y) * 2 

x = Rectangle(2,2)
print(f"Rectangle Square: {x.square()}")
print(f"Rectangle Perimeter: {x.perimeter()}")
```

## # 2 Use Except -> try-except

```python
def safe_division():
    try:
        numerator_str = input("Бөлінгіш санды (алушыны) енгізіңіз: ")
        denominator_str = input("Бөлгіш санды енгізіңіз: ")

        numerator = float(numerator_str)
        denominator = float(denominator_str)

        result = numerator / denominator

    except ValueError: 
        print("\n Қате: Дұрыс санды енгізіңіз (мысалы, 5 немесе 2.5). Мәтін енгізуге болмайды.")
    except ZeroDivisionError:
        print("\n Қате: Кез келген санды нөлге бөлуге болмайды.")
    else:
        print(f"\n✅ Нәтиже: {numerator} / {denominator} = {result}")
    finally:
        print("Бөлу операциясы аяқталды.")

safe_division()
```

# # 3 My math modul

```python
import mymath 
```

```python
num1 = 15
num2 = 5
num3 = 0

print(f"Қосу: {num1} + {num2} = {mymath.add(num1, num2)}")
print(f"Азайту: {num1} - {num2} = {mymath.sub(num1, num2)}")
print(f"Көбейту: {num1} * {num2} = {mymath.mul(num1, num2)}")

try:
    print(f"Бөлу: {num1} / {num2} = {mymath.div(num1, num2)}")
    print(f"Нөлге бөлуді тексеру: {num1} / {num3} = {mymath.div(num1, num3)}")
except ZeroDivisionError as e:
    print(f"Бөлу қатесі: {e}")
```
