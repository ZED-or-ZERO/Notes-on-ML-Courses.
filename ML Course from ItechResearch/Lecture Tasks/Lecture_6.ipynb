```python
# Task 2 

data = {
    'name': ['Aigerim', 'Dias', 'Miras', 'Dana'],
    'age': [20, 22, 21, 19],
    'score': [88, 92, 75, 95]
}

students_df = pd.DataFrame(data)
print("Students DataFrame:")
print(students_df)

older_students = students_df[students_df['age'] > 20]
print("\n Students where age more 20 years:")
print(older_students)

name_score = students_df[['name', 'score']]
print("\n'name' and 'score' columns:")
print(name_score)

max_score = students_df['score'].max()
top_student_name = students_df[students_df['score'] == max_score]['name'].iloc[0]
print(f"\n Max Score: {top_student_name} ({max_score})")
```

```python
#               Task 3 

value_iloc = students_df.iloc[2, 2]
print(value_iloc)

dias_row = students_df.loc[students_df['name'] == 'Dias']
print("\nName 'Dias' students:")
print(dias_row)
```

```python
#               Task 4

drugs_data = {
    'drug': ['A', 'B', 'C', 'D'],
    'price': [1200, 800, 600, 2000],
    'amount': [5, 10, 3, 8]
}
drugs_df = pd.DataFrame(drugs_data)
print("Dariler DataFrame:")
print(drugs_df)

# 1 
expensive_drugs = drugs_df[drugs_df['price'] > 1000]
print("\n Price more 1000:")
print(expensive_drugs)

# 2
price_range_drugs = drugs_df[(drugs_df['price'] >= 600) & (drugs_df['price'] <= 800)]
print("\nPrice between 600 and 800:")
print(price_range_drugs)

# 3
low_amount_drugs = drugs_df[drugs_df['amount'] < 5]
print("\nPrice less 5:")
print(low_amount_drugs)
```

```python
#               Task 5

employees_data = {
    'name': ['Ayan', 'Dana', 'Miras', 'Aida', 'Timur'],
    'department': ['IT', 'HR', 'IT', 'HR', 'Sales'],
    'salary': [400000, 300000, 450000, 320000, 280000]
}
```

```python
employees_df = pd.DataFrame(employees_data)
employees_df 
```

```python
# 1 
avg_salary_by_dept = employees_df.groupby('department')['salary'].mean()
print("\nӘр бөлім бойынша орташа жалақы:")
print(avg_salary_by_dept)

# 2
count_by_dept = employees_df.groupby('department')['name'].count()
print("\nӘр бөлімдегі қызметкерлер саны:")
print(count_by_dept)

# 3
highest_avg_salary_dept = avg_salary_by_dept.idxmax()
print(f"\nЕң жоғары орташа жалақысы бар бөлім: {highest_avg_salary_dept} ({avg_salary_by_dept.max():.0f} тг)")
```

```python
#                   Task 6


min_max_salary = employees_df.groupby('department')['salary'].agg(['min', 'max'])
print("\nӘр бөлім үшін жалақының min және max мәндері:")
print(min_max_salary)

total_salary_by_dept = employees_df.groupby('department')['salary'].sum()
print("\nӘр бөлімдегі жалақының жалпы суммасы:")
print(total_salary_by_dept)
```

```python
#               Task 7 

students = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Aigerim', 'Dias', 'Ayan']
})

scores = pd.DataFrame({
    'id': [1, 2, 3],
    'math': [88, 75, 95],
    'physics': [90, 80, 92]
})

print("students DataFrame:")
print(students)
print("\nscores DataFrame:")
print(scores)

# 1     
merged_df = pd.merge(students, scores, on='id')
print(merged_df)

# 2
students_indexed = students.set_index('id')
scores_indexed = scores.set_index('id')
joined_df = students_indexed.join(scores_indexed)
print(joined_df)

# 3
concatenated_rows = pd.concat([students, students], ignore_index=True)
print(concatenated_rows)

# 4
concatenated_cols = pd.concat([students, scores], axis=1)
print(concatenated_cols)
```

```python
#                   Task 8

advanced_data = {
    'student': ['Aigerim', 'Dias', 'Ayan'],
    'math': [88, 75, 95],
    'physics': [90, 80, 92],
    'history': [78, 85, 90]
}
grades_df = pd.DataFrame(advanced_data)
grades_df = grades_df.set_index('student') 
print("Бастапқы бағалар DataFrame:")
print(grades_df)

subject_stats = grades_df.agg(['min', 'max', 'mean', 'median'])
print("\nӘр пән бойынша жалпы статистика:")
print(subject_stats)

grades_df['avg'] = grades_df.mean(axis=1)
print("\n'avg' бағаны қосылған DataFrame:")
print(grades_df)

high_achievers = grades_df[grades_df['avg'] > 85]
print("\nОрташа балы 85-тен жоғары студенттер:")
print(high_achievers)
```
